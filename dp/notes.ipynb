#include <bits/stdc++.h>
class Solution {
  public:
    int fun(int i,int maxor,int ans,int n,vector<int> &arr,vector<vector<int>> &dp)
    {
        if(maxor==ans)
        {
            return 0;
        }
        if(i==n)
        {
            return 1e9;
        }
        if(dp[i][maxor]!=-1)
        {
            return dp[i][maxor];
        }
        int updatedMaxor=(maxor| arr[i]);
        return dp[i][maxor]=min(fun(i+1,maxor,ans,n,arr,dp),1+fun(i+1,updatedMaxor,ans,n,arr,dp));
    }
    int minLength(vector<int>& arr) {
        int n = arr.size();
        int ans=0;
        vector<vector<int>> dp(n+1,vector<int>(10000,-1));
        for(int i=0;i<n;i++)
        {
            ans|=arr[i];   
        }
        
        return fun(0,0,ans,n,arr,dp);
    }
};

// striver code
// Function to find the length of the longest increasing subsequence
int getAns(int arr[], int n, int ind, int prev_index, vector<vector<int>>& dp) {
    // Base condition
    if (ind == n)
        return 0;
        
    if (dp[ind][prev_index + 1] != -1)
        return dp[ind][prev_index + 1];
    
    int notTake = 0 + getAns(arr, n, ind + 1, prev_index, dp);
    
    int take = 0;
    
    if (prev_index == -1 || arr[ind] > arr[prev_index]) {
        take = 1 + getAns(arr, n, ind + 1, ind, dp);
    }
    
    return dp[ind][prev_index + 1] = max(notTake, take);
}

int longestIncreasingSubsequence(int arr[], int n) {
    // Create a 2D DP array initialized to -1
    vector<vector<int>> dp(n, vector<int>(n + 1, -1));
    
    return getAns(arr, n, 0, -1, dp);
}

<div style="text-align:center; font-size:24px; font-weight:bold; margin-top:20px; color:blue;">How to Determine the State of DP</div>
<ol>
<li>
Understand the Goal
Ask yourself: What am I trying to compute?
In this problem, it's the number of valid sequences ending at index i with specific values.
</li>

<li>


Look for Patterns
Think about how a solution at index i depends on previous results.
Example: If arr[i] == 0, all values from 1 to m are possible. Each value depends on transitions from i-1.
</li>

<li>


Define the Smallest Subproblem
Break the problem into the smallest meaningful unit.
Example: dp[i][j] could represent the number of ways to fill the sequence up to index i where the i-th element is j.
</li>

<li>

Think Recursively
Imagine calculating for one step:

If arr[i] == 0: The value j can come from j, j-1, or j+1 at i-1.
If arr[i] != 0: Only transitions leading to arr[i] matter.
Test Hypotheses on Small Inputs
Take a small example, like n=3, m=3, arr=[0, 0, 0], and manually calculate outcomes to validate your state definition.
Example: At i=2, j=2, manually trace transitions from i=1.
</li>


</ol>

	int arr[n];
	for (int i = 0; i < n; i++) { cin >> arr[i]; }
	if (arr[0] == 0) {
		fill(dp[0], dp[0] + 101, 1);
	} else {
		dp[0][arr[0]] = 1;
	}
	for (int i = 1; i < n; i++) {
		if (arr[i] == 0) {
			for (int j = 1; j <= m; j++) {
				dp[i][j] += dp[i - 1][j];
				if (j - 1 > 0) dp[i][j] += dp[i - 1][j - 1];
				if (j + 1 <= m) dp[i][j] += dp[i - 1][j + 1];
				dp[i][j] %= MOD;
			}
		} else {
			dp[i][arr[i]] += dp[i - 1][arr[i]];
			if (arr[i] - 1 > 0) dp[i][arr[i]] += dp[i - 1][arr[i] - 1];
			if (arr[i] + 1 <= m) dp[i][arr[i]] += dp[i - 1][arr[i] + 1];
			dp[i][arr[i]] %= MOD;
		}
	}
	ll ans = 0;
	for (int i = 1; i <= m; i++) {
		ans += dp[n - 1][i];
		ans %= MOD;
	}
